## Task 1



- The original data has 1500 rows and 8 columns.
- product_id contains 1500 unique values. No missing values were found.
- There were 5 distinct categories and 25 missing values ("-") that were replaced with "Unknown".
- There were 4 unique animals as expected.
- The size column had 9 distinct values, all variations of "Small", "Medium", "Large". I used string functions to clean the data and reduce the total to the three categories as requested.
- In the price column, there were 150 missing values ("unlisted"). I utilized CASE statements and MEDIAN() function in a nested query to replace them with the overall median price as requested. All values were rounded to 2 decimal places.
- All values in the sales column were rounded to 2 decimal plates. There were no missing values.
- In the rating column, there was one missing value (NA) that was replaced with 0 as requested.
- There were no null values in repeat_purchase.


```python
--Original table
SELECT * FROM 'pet_supplies_2212.csv'
```

```python
--CLEAN TABLE
SELECT
product_id,
CASE WHEN category = '-' THEN 'Unknown'
	ELSE category
	END AS category,
animal,
LOWER(size) AS size,
CASE WHEN price = 'unlisted' THEN (SELECT median(price) FROM 'pet_supplies_2212.csv')
	ELSE price 
	END AS price,
ROUND(sales,2) AS sales,
CASE WHEN rating = 'NA' then '0'
	ELSE rating
	END AS rating,
repeat_purchase
FROM 'pet_supplies_2212.csv'
```
